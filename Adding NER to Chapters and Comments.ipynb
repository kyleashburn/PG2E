```python
import pandas as pd
import spacy
from spacy import displacy
from spacy import tokenizer

nlp = spacy.load("en_core_web_trf") # loading in the transformer model we're working with here
spacy.prefer_gpu() # setting spacy to use the gpu to accellerate computation here
```

```python
# reading in the chapter data we're working with 
chapters = pd.read_json("book_chapters.jl", lines=True)
```

```python
chapters
```

```python
%%time 
chapt_ner = list()
for doc in nlp.pipe(chapters["chapt_text"], batch_size=50, 
                              disable=["tok2vec", "tagger", "parser", "attribute_ruler", "lemmatizer"]):
    chapt_ner.append([(ent.text, ent.label_) for ent in doc.ents])
```

```python
chapters["chapt_ner"] = chapt_ner # making the new series with the thingy I need there
```

```python
chapters.to_json("book_chapters_w_NER.jl", orient="records", lines=True) # saving our work
```
